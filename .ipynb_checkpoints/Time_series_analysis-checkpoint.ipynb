# Time Series Analysis of Racing Game Input Data

In this notebook, we will perform a time series analysis on the gameplay input data from a racing car game. The data will be read from a text file, processed, and analyzed to identify patterns in the accelerator, brake, and left/right control inputs.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

## Step 1: Read and Parse the Input Log File

We will read the input log file and parse it to extract timestamps along with the corresponding values for acceleration, braking, and left/right controls.

```python
# Step 1: Read and parse the input log file
file_path = 'input_log.txt'  # Update with your file path if necessary
data = []

with open(file_path, 'r') as f:
    for line in f:
        parts = line.strip().split(' at ')
        if len(parts) == 2:
            value_info = parts[0]
            timestamp_str = parts[1]
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S.%f")
            
            if 'Accelerator' in value_info:
                value = float(value_info.split(' moved to ')[1])
                data.append((timestamp, 'Accelerator', value))
            elif 'Brake' in value_info:
                value = float(value_info.split(' moved to ')[1])
                data.append((timestamp, 'Brake', value))
            elif 'Left/Right Control' in value_info:
                value = float(value_info.split(' moved to ')[1])
                data.append((timestamp, 'Left/Right Control', value))
```

## Step 2: Create a DataFrame

Next, we will create a Pandas DataFrame from the parsed data for easier manipulation and analysis.

```python
# Step 2: Create a DataFrame
df = pd.DataFrame(data, columns=['Timestamp', 'Control', 'Value'])
```

```python
df.head(10)
```

```python
# Step 3: Aggregate duplicate entries by taking the mean
df_grouped = df.groupby(['Timestamp', 'Control']).mean().reset_index()
```

## Step 3: Pivot the DataFrame for Easier Plotting

We will pivot the DataFrame so that each control type becomes a separate column with timestamps as indices.

```python
pivot_df = df_grouped.pivot(index='Timestamp', columns='Control', values='Value')
```

```python
pivot_df.head(10)
```

## Step 4: Plotting the Data

Now we will plot the values of accelerator, brake, and left/right controls over time to visualize their behavior.

```python
plt.figure(figsize=(25, 8))
plt.plot(pivot_df.index, pivot_df['Accelerator'], label='Accelerator (Axis 5)', color='g')
plt.plot(pivot_df.index, pivot_df['Brake'], label='Brake (Axis 4)', color='r')
plt.plot(pivot_df.index, pivot_df['Left/Right Control'], label='Left/Right Control (Axis 0)', color='b')
plt.title('Gameplay Input Analysis Over Time')
plt.xlabel('Time')
plt.ylabel('Control Value')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Step 5: Analyze Patterns (Optional)

We can also calculate correlations between different controls to examine their relationships.

```python
# Step 5: Analyze patterns (optional)
correlation_matrix = pivot_df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Optional: Classify patterns based on thresholds or conditions
def classify_patterns(row):
    if row['Accelerator'] > 0.5 and row['Brake'] < -0.5:
        return 'Aggressive Acceleration'
    elif row['Brake'] > 0.5:
        return 'Braking'
    elif row['Left/Right Control'] < -0.5:
        return 'Turning Left'
    elif row['Left/Right Control'] > 0.5:
        return 'Turning Right'
    else:
        return 'Neutral'

pivot_df['Pattern'] = pivot_df.apply(classify_patterns, axis=1)

# Display classified patterns
print(pivot_df[['Pattern']].head(100))
```

```python
# Step 6: Visualize patterns over time
plt.figure(figsize=(14, 8))
plt.plot(pivot_df.index, pivot_df['Pattern'].astype('category').cat.codes, label='Driving Pattern', marker='o')

# Customize ticks and labels
plt.yticks(ticks=range(len(pivot_df['Pattern'].unique())), labels=pivot_df['Pattern'].unique())
plt.title('Driving Patterns Over Time')
plt.xlabel('Time Index')
plt.ylabel('Driving Pattern')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

## #5 Advnace 

```python
import numpy as np
# Step 5: Define a function to analyze patterns within a running window
def analyze_window(data_window):
    features = {
        'mean_accelerator': np.mean(data_window['Accelerator']),
        'mean_brake': np.mean(data_window['Brake']),
        'mean_turning': np.mean(data_window['Left/Right Control']),
        'std_accelerator': np.std(data_window['Accelerator']),
        'std_brake': np.std(data_window['Brake']),
        'std_turning': np.std(data_window['Left/Right Control'])
    }
    return features
```

```python
# Step 6: Sliding window parameters
window_size = 5  # Number of rows in each window
step_size = 1    # Number of rows to move for each step
```

```python
# Step 7: Analyze each window and classify patterns
results = []
for start in range(0, len(pivot_df) - window_size + 1, step_size):
    end = start + window_size
    window_data = pivot_df.iloc[start:end]
    features = analyze_window(window_data)
    
    # Classify patterns based on features (you can adjust thresholds)
    if features['mean_accelerator'] > 0.5 and features['mean_brake'] < 0:
        pattern = 'Aggressive Acceleration'
    elif features['mean_brake'] > 0.5:
        pattern = 'Braking'
    elif features['mean_turning'] < -0.5:
        pattern = 'Turning Left'
    elif features['mean_turning'] > 0.5:
        pattern = 'Turning Right'
    else:
        pattern = 'Neutral'
    
    results.append((pivot_df.index[start], features, pattern))
```

```python
# Convert results to DataFrame for easier analysis
results_df = pd.DataFrame(results, columns=['Start_Index', 'Features', 'Pattern'])

# Display some results
print(results_df[['Start_Index', 'Pattern']].head(50))
```

## 1. Statistical Analysis to Identify Time Patterns

```python
import statsmodels.api as sm

# Check for autocorrelation patterns in each control type
def plot_autocorrelation(data, control_type):
    series = data[control_type].dropna()
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
    sm.graphics.tsa.plot_acf(series, lags=40, ax=ax1)
    sm.graphics.tsa.plot_pacf(series, lags=40, ax=ax2)
    ax1.set_title(f'Autocorrelation - {control_type}')
    ax2.set_title(f'Partial Autocorrelation - {control_type}')
    plt.show()

# Decompose the time series into trend, seasonal, and residuals
def plot_decomposition(data, control_type):
    series = data[control_type].dropna()
    decomposition = sm.tsa.seasonal_decompose(series, model='additive', period=24)  # Adjust period based on data frequency
    decomposition.plot()
    plt.suptitle(f'Time Series Decomposition - {control_type}', fontsize=16)
    plt.show()

# Apply to each control type
for control in ['Accelerator', 'Brake', 'Left/Right Control']:
    plot_autocorrelation(pivot_df, control)
    plot_decomposition(pivot_df, control)
```

### 1. Calculate and Print Autocorrelation (ACF) and Partial Autocorrelation (PACF)

```python
from statsmodels.tsa.stattools import acf, pacf

def calculate_acf_pacf(data, control_type, max_lag=20):
    # Drop NaN values for the control type
    series = data[control_type].dropna()
    
    # Calculate ACF and PACF
    acf_values = acf(series, nlags=max_lag)
    pacf_values = pacf(series, nlags=max_lag)
    
    # Print ACF and PACF values
    print(f"\nAutocorrelation (ACF) values for {control_type} (up to lag {max_lag}):")
    for lag, value in enumerate(acf_values):
        print(f"Lag {lag}: {value:.3f}")
    
    print(f"\nPartial Autocorrelation (PACF) values for {control_type} (up to lag {max_lag}):")
    for lag, value in enumerate(pacf_values):
        print(f"Lag {lag}: {value:.3f}")

# Apply to each control type
for control in ['Accelerator', 'Brake', 'Left/Right Control']:
    calculate_acf_pacf(pivot_df, control)
```

### 2.Seasonal-Trend Decomposition (STL) and Printing Components

```python
from statsmodels.tsa.seasonal import STL

def decompose_and_print(data, control_type):
    series = data[control_type].dropna()
    stl = STL(series, seasonal=7)  # Set seasonal period according to data pattern
    
    # Perform decomposition
    result = stl.fit()
    
    # Print decomposed components (Trend, Seasonal, Residual)
    print(f"\nSTL Decomposition for {control_type}:")
    print("Trend Component (first 10 values):")
    print(result.trend.head(10))
    print("\nSeasonal Component (first 10 values):")
    print(result.seasonal.head(10))
    print("\nResidual Component (first 10 values):")
    print(result.resid.head(10))
    
    # Optionally, plot the decomposition
    result.plot()
    plt.suptitle(f'STL Decomposition - {control_type}', fontsize=16)
    plt.show()

# Apply STL decomposition to each control type
for control in ['Accelerator', 'Brake', 'Left/Right Control']:
    decompose_and_print(pivot_df, control)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler

def prepare_data(series, n_lags=1):
    # Convert series to supervised learning format
    X, y = [], []
    for i in range(n_lags, len(series)):
        X.append(series[i-n_lags:i])
        y.append(series[i])
    return np.array(X), np.array(y)

# Assuming pivot_df is already defined with the 'Accelerator' column
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(pivot_df[['Accelerator']].values)

# Prepare data with lags
n_lags = 3  # Using a week of history
X, y = prepare_data(scaled_data, n_lags)

# Check if X and y have valid shapes
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

if X.shape[0] == 0:
    print("Not enough data to create sequences with n_lags =", n_lags)
else:
    # Reshape X for LSTM model (samples, timesteps, features)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Define LSTM model
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(n_lags, 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')

    # Train the model
    model.fit(X, y, epochs=20, verbose=1)

    # Forecast multiple future values
    n_forecast_steps = 10
    forecast = []

    # Use the last n_lags steps as the initial input
    forecast_input = scaled_data[-n_lags:].reshape((1, n_lags, 1))

    for _ in range(n_forecast_steps):
        yhat = model.predict(forecast_input)  # Predict next value
        forecast.append(yhat[0, 0])           # Store the prediction
        
        # Reshape yhat to (1, 1, 1) for concatenation (no need for expand_dims)
        yhat_reshaped = yhat.reshape(1, 1, 1)
        
        # Update the input by concatenating the predicted value and removing the oldest
        forecast_input = np.concatenate((forecast_input[:, 1:, :], yhat_reshaped), axis=1)

    # Inverse transform the forecasted values to the original scale
    forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

    # Print the forecasted values
    print("Forecasted values:", forecast)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler

# Prepare data function (same as before)
def prepare_data(series, n_lags=1):
    X, y = [], []
    for i in range(n_lags, len(series)):
        X.append(series[i-n_lags:i])
        y.append(series[i])
    return np.array(X), np.array(y)

# Function to train and forecast for each category
def train_and_forecast(category_data, n_lags=3, n_forecast_steps=10):
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(category_data.values.reshape(-1, 1))
    
    # Prepare data with lags
    X, y = prepare_data(scaled_data, n_lags)
    
    # Check if X and y have valid shapes
    print("Shape of X:", X.shape)
    print("Shape of y:", y.shape)
    
    if X.shape[0] == 0:
        print("Not enough data to create sequences with n_lags =", n_lags)
    else:
        # Reshape X for LSTM model (samples, timesteps, features)
        X = X.reshape((X.shape[0], X.shape[1], 1))

        # Define LSTM model
        model = Sequential([
            LSTM(50, activation='relu', input_shape=(n_lags, 1)),
            Dense(1)
        ])
        model.compile(optimizer='adam', loss='mse')

        # Train the model
        model.fit(X, y, epochs=20, verbose=1)

        # Forecast multiple future values
        forecast = []

        # Use the last n_lags steps as the initial input
        forecast_input = scaled_data[-n_lags:].reshape((1, n_lags, 1))

        for _ in range(n_forecast_steps):
            yhat = model.predict(forecast_input)  # Predict next value
            forecast.append(yhat[0, 0])           # Store the prediction

            # Reshape yhat to (1, 1, 1) for concatenation
            yhat_reshaped = yhat.reshape(1, 1, 1)

            # Update the input by concatenating the predicted value and removing the oldest
            forecast_input = np.concatenate((forecast_input[:, 1:, :], yhat_reshaped), axis=1)

        # Inverse transform the forecasted values to the original scale
        forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

        # Print the forecasted values
        return forecast

# Assuming pivot_df is already defined with 'Accelerator', 'Category1', 'Category2' columns
forecast_accelerator = train_and_forecast(pivot_df['Accelerator'], n_lags=3, n_forecast_steps=10)
forecast_category1 = train_and_forecast(pivot_df['Brake'], n_lags=3, n_forecast_steps=10)
forecast_category2 = train_and_forecast(pivot_df['Left/Right Control'], n_lags=3, n_forecast_steps=10)

# Print the forecasted values for each category
print("Forecasted values for Accelerator:", forecast_accelerator)
print("Forecasted values for Category1:", forecast_category1)
print("Forecasted values for Category2:", forecast_category2)
```
